```python
df = pd.read_csv(filepaths[0], delimiter=',',encoding='ISO-8859-1',low_memory=False)
```

```python
df.head()
```

```python
df.rename(columns={'Unnamed: 0':'id'}, inplace=True)
```

```python
stdates=[]
for i in range(len(df)):
    stdate = datetime.strptime(str(df['styear'][i])+'-'+str(df['stmonth'][i])+'-'+str(df['stday'][i]),'%Y-%m-%d').date()
    stdates.append(stdate)
```

```python
df['stdate'] = stdates
```
