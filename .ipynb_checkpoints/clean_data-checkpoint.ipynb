```python
import pandas as pd
```

```python
df = pd.read_fwf('Appendix_A.txt', header=0)
df.head()
```

```python
dflist = df['Year, War Name, War Type & Number'].to_list()
Year = []
WarName = []
WarType = []
WarNumber = []

for i in range(len(dflist)):
    words = dflist[i].split()
    # print(words)
    year = words[0]
    warNumber = words[-1]
    warType = str(words[-3:-2])
    warType = warType.strip('[\']')
    warType += ' War'
    warName = ''
    for n in range(len(words)-4):
        r = n + 1
        warName += str(words[r] + ' ')
    warName = warName.strip()
    Year.append(year)
    WarName.append(warName)
    WarType.append(warType)
    WarNumber.append(warNumber)
    print(year)
    print(warName)
    print(warType)
    print(warNumber + '\n')
    
```

```python
dict = {'Year':Year,'War Name':WarName,'War Type':WarType,'War Number':WarNumber}
df = pd.DataFrame(dict)
df.head()
```

```python
df.to_csv('appendix_a.csv')
```
